```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Aragon')
```

```python
_ = plt.pie(fracs, labels=labels)
```

```python
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, labels=labels)
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels)
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels, shadow=True)
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels, autopct='%2.0f %%')
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels, autopct='%2.0f %%', pctdistance=0.9)
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels, autopct='%2.0f %%')
```

```python
explode=(0, 0.2, 0)
plt.gca().set_aspect('equal')
_ = plt.pie(fracs, explode=explode, labels=labels, autopct='%2.0f %%')
_ = plt.table(cellText=[fracs], colLabels=labels, rowLabels=['Fraction'], loc='top') 
```
