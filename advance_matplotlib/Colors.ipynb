## Color Maps

```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy.ndimage.filters import gaussian_filter
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
g_array = gaussian_filter(np.random.normal(size=(200, 300)), sigma=10)
```

```python
plt.imshow(g_array)
```

```python
plt.imshow(g_array, cmap='inferno')
```

```python
plt.imshow(g_array, cmap='seismic')
```

```python
plt.imshow(g_array, cmap='flag')
```

```python
plt.imshow(g_array, cmap=mpl.colors.ListedColormap(['r', 'b', 'g']))
plt.colorbar()
```

```python
cdict = dict(red=[(0, 0, 1), (0.5, 1, 1), (1, 1, 1)],
             green=[(0, 0, 1), (1, 0, 1)],
             blue=[(0, 0, 1), (1, 0, 1)])

plt.imshow(g_array, cmap=mpl.colors.LinearSegmentedColormap('mymap', cdict))
plt.colorbar()
```

## Color Cycling

```python
from cycler import cycler
```

```python
nums = np.arange(0, 10, 0.1)
```

```python
plt.plot(nums, np.sin(nums), label='1')
plt.plot(1+nums, np.sin(nums), label='2')
plt.plot(2+nums, np.sin(nums), label='3')
plt.plot(3+nums, np.sin(nums), label='4')
plt.plot(4+nums, np.sin(nums), label='5')
plt.legend(loc='best')
```

```python
plt.plot(nums, np.sin(nums), label='1')
plt.plot(1+nums, np.sin(nums), label='2')
plt.gca().set_prop_cycle(None)
plt.plot(2+nums, np.sin(nums), label='3')
plt.plot(3+nums, np.sin(nums), label='4')
plt.gca().set_prop_cycle(None)
plt.plot(4+nums, np.sin(nums), label='5')
plt.legend(loc='best')
```

```python
color_cyc = cycler('color', ['red', 'blue', 'green'])
plt.gca().set_prop_cycle(color_cyc)

plt.plot(nums, np.sin(nums), label='1')
plt.plot(1+nums, np.sin(nums), label='2')
plt.plot(2+nums, np.sin(nums), label='3')
plt.plot(3+nums, np.sin(nums), label='4')
plt.plot(4+nums, np.sin(nums), label='5')
plt.legend(loc='best')
```
