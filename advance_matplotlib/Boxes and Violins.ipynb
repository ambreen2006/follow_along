```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
rand1 = np.random.normal(size=500)
rand2 = np.random.normal(size=500, scale=2)
rand3 = np.random.normal(size=500, scale=0.5, loc=1)
```

```python
gaussians = (rand1, rand2, rand3)
```

## Box Plot

```python
_ = plt.boxplot(gaussians)
```

```python
_ = plt.boxplot(gaussians, labels=("first", "second", "third"))
```

```python
_ = plt.boxplot(gaussians, labels=("first", "second", "third"), vert=False)
```

## Violin Plots

```python
rand4 = np.random.lognormal(size=500, sigma=0.5)
rand5 = np.random.pareto(size=500, a=5)
```

```python
dists = (rand1, rand4, rand5)
```

```python
_ = plt.violinplot(dists)
```

```python
_ = plt.violinplot(dists, showmedians=True, showmeans=True)
```

```python
vplot = plt.violinplot(dists, showmedians=True, showmeans=True)
vplot['cmedians'].set_color(['b', 'g', 'k'])
vplot['bodies'][0].set_color('k')
vplot['bodies'][1].set_color('g')
```
