## Plotting lines

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
def basic_sin():
    nums = np.arange(0, 10, 0.1)
    plt.plot(nums, np.sin(nums))
```

```python
basic_sin()
```

```python
plt.axhline(0, color='r')
plt.axvline(np.pi/2., color='r', linestyle='--')
basic_sin()
```

```python
plt.axhspan(-0.5, 0.5, alpha=0.2)
basic_sin()
```

```python
plt.axhspan(-0.5, 0.5, xmin=0.5, xmax=0.75, alpha=0.2)
basic_sin()
```

From the documentation:

xminfloat, default: 0
Lower x-coordinate of the span, in x-axis (0-1) units.

xmaxfloat, default: 1
Upper x-coordinate of the span, in x-axis (0-1) units.

```python
plt.grid(color='r', linestyle=':')
basic_sin()
```

```python
plt.grid(color='r', linestyle='--')
basic_sin()
```

## Polygons and Other Shapes

```python
circle = mpl.patches.Circle((5,0), radius=1)
plt.gca().add_patch(circle)
basic_sin()
```

```python
circle = mpl.patches.Circle((5,0), radius=1)
plt.gca().add_patch(circle)
plt.gca().set_aspect('equal')
basic_sin()
```

```python
plt.gca().set_aspect('equal')

c1 = mpl.patches.Circle((5,0), radius=1)
c2 = mpl.patches.Circle((4,0), radius=1, color='r')

plt.gca().add_patch(c1)
plt.gca().add_patch(c2)


basic_sin()
```

```python
plt.gca().set_aspect('equal')

c1 = mpl.patches.Circle((5,0), radius=1, zorder=1)
c2 = mpl.patches.Circle((4,0), radius=1, color='r', zorder=0)

plt.gca().add_patch(c1)
plt.gca().add_patch(c2)


basic_sin()
```

```python
p = mpl.patches
nums = np.arange(0, 10, 0.1)

circle = p.Circle(xy=(2, -2), radius=1)
arc = p.Arc(xy=(1,2), width=1, height=3, angle=0, theta1=90, theta2=270)
wedge = p.Wedge(center=(2, 2), r=1, theta1=-180, theta2=100, width=0.5)
arrow = p.Arrow(x=5, y=-3, dx=2, dy=2)
ellipse = p.Ellipse(xy=(5, 2), width=1, height=3, angle=60)
rect = p.Rectangle(xy=(7, -2), width=2, height=2, angle=-30)
poly = p.RegularPolygon(xy=(8, 2), numVertices=3, orientation=45, radius=1)

plt.plot(nums, 10/3. * np.sin(nums))

plt.gca().add_patch(circle)
plt.gca().add_patch(arc)
plt.gca().add_patch(wedge)
plt.gca().add_patch(arrow)
plt.gca().add_patch(ellipse)
plt.gca().add_patch(rect)
plt.gca().add_patch(poly)
```

### Making your own polygons

```python
pos = [(3, 0), (7, 0), (4, 1), (6, 1)]
poly = mpl.patches.Polygon(pos)

plt.plot(nums, 10/3. * np.sin(nums))
plt.gca().add_patch(poly)
```

```python
pos = [(3, 0), (7, 0), (6, 1), (4, 1)]
poly = mpl.patches.Polygon(pos, color='purple', zorder=1)

plt.plot(nums, 10/3. * np.sin(nums), zorder=0)
plt.gca().add_patch(poly)
```

## Annotations

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(facecolor='k'))
basic_sin()
```

```python
plt.annotate("Sine Wave", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(facecolor='k'))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(facecolor='k', shrink=0.5, 
                                                                 headwidth=50, headlength=5))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(arrowstyle='fancy'))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(arrowstyle='|-|'))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(arrowstyle='fancy, head_length=2, head_width=2'))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(arrowstyle='fancy', 
                                                                 connectionstyle='arc3, rad=0.5'))
basic_sin()
```

```python
plt.annotate("", xy=(np.pi/2, 1), xytext=(5, 0), arrowprops=dict(arrowstyle='fancy', 
                                                                 connectionstyle='angle3, angleA=-30, angleB=60'))
basic_sin()
```
