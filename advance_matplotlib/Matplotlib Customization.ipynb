```python
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

from scipy.ndimage.filters import gaussian_filter
```

```python
%matplotlib inline
```

## Using Different Styles

```python
plt.style.available
```

```python
plt.style.use("dark_background")

plt.subplot(221)
plt.plot(np.arange(0, 1, 0.01), np.power(np.arange(0, 1, 0.01), 3))
plt.ylabel('Axis Label')

plt.subplot(222)
plt.scatter(np.random.normal(size=10), np.random.normal(size=10), c=np.random.normal(size=10))

plt.subplot(223)
plt.hist(np.random.normal(size=1000))
plt.hist(np.random.normal(1, size=1000))
plt.hist(np.random.normal(2, size=1000))

plt.ylabel('Axis Label')
plt.xlabel('Axis Label')

plt.subplot(224)
plt.imshow(gaussian_filter(np.random.normal(size=(200, 300)), sigma=10))
plt.xlabel('Axis Label')
```

```python
plt.style.use('classic')

plt.subplot(221)
plt.plot(np.arange(0, 1, 0.01), np.power(np.arange(0, 1, 0.01), 3))
plt.ylabel('Axis Label')

plt.subplot(222)
with plt.style.context('seaborn-dark'):
    plt.scatter(np.random.normal(size=10), np.random.normal(size=10), c=np.random.normal(size=10))

plt.subplot(223)
plt.hist(np.random.normal(size=1000))
plt.hist(np.random.normal(1, size=1000))
plt.hist(np.random.normal(2, size=1000))

plt.ylabel('Axis Label')
plt.xlabel('Axis Label')

plt.subplot(224)
plt.imshow(gaussian_filter(np.random.normal(size=(200, 300)), sigma=10))
plt.xlabel('Axis Label')
```

## Creating Custom Style Templates

Guide is available @ https://matplotlib.org/3.3.3/tutorials/introductory/customizing.html

```python
mpl.get_configdir()
```

Create file with the extension `mplstyle`

Place any style customization in it, the file being used here is called `bigpoints.mplstyle`

`lines.markersize:50`

Then place this file in the `<configdir>/stylelib`

```python
plt.style.available
```

```python
plt.style.use('bigpoints')

plt.subplot(221)
plt.plot(np.arange(0, 1, 0.01), np.power(np.arange(0, 1, 0.01), 3))
plt.ylabel('Axis Label')

plt.subplot(222)
with plt.style.context('seaborn-dark'):
    plt.scatter(np.random.normal(size=10), np.random.normal(size=10), c=np.random.normal(size=10))

plt.subplot(223)
plt.hist(np.random.normal(size=1000))
plt.hist(np.random.normal(1, size=1000))
plt.hist(np.random.normal(2, size=1000))

plt.ylabel('Axis Label')
plt.xlabel('Axis Label')

plt.subplot(224)
plt.imshow(gaussian_filter(np.random.normal(size=(200, 300)), sigma=10))
plt.xlabel('Axis Label')
```

## Mixing Styles

To use more than one style, pass an array to plt.style.use
