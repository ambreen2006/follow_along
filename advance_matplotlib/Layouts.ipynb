```python
plt.subplot(131)
plt.subplot(132)
plt.subplot(133)
```

```python
plt.subplot(221)
plt.subplot(222)
plt.subplot(224)
```

```python
plt.subplot2grid((2, 2), (0, 0))
plt.subplot2grid((2, 2), (1, 1))
plt.subplot2grid((2, 2), (1, 0))
```

```python
plt.subplot2grid((2, 2), (0, 0), colspan=2)
plt.subplot2grid((2, 2), (1, 1))
plt.subplot2grid((2, 2), (1, 0))
```

```python
gs = mpl.gridspec.GridSpec(2, 2)
plt.subplot(gs[0, 0])
plt.subplot(gs[0, 1])
```

```python
gs = mpl.gridspec.GridSpec(2, 2)
plt.subplot(gs[0, :])
plt.subplot(gs[1, 1])
```

```python
gs = mpl.gridspec.GridSpec(2, 3)
plt.subplot(gs[0, 1:])
plt.subplot(gs[1, 1])
```

```python
gs = mpl.gridspec.GridSpec(2, 2, height_ratios=[0.5, 2])
plt.subplot(gs[0, 0])
plt.subplot(gs[0, 1])
plt.subplot(gs[1, 0])
plt.subplot(gs[1, 1])
```
