```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
x = np.random.normal(size=500)
y = np.random.normal(size=500)
z = np.random.normal(size=500)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=np.linalg.norm([x, y, z], axis=0))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x, y = np.meshgrid(np.arange(-10, 10, 0.5), np.arange(-10, 10, 0.5))
r = np.linalg.norm([x, y], axis=0)

goldstone = -160 * np.power(r, 2) * np.power(r, 4)
ax.scatter(x, y, goldstone)
```

```python
# Not from the tutorial but I wanted to do this

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = X + Y

ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.8, cmap=mpl.cm.ocean)
```
