```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (6, 4)
plt.rcParams['figure.dpi'] = 150
```

```python
nums = np.arange(0, 10, 0.1)
```

```python
plt.plot(nums, nums/10.)
plt.plot(nums, np.sin(nums))
plt.plot(nums, np.cos(nums))
```

```python
plt.subplot(111, projection='polar')
plt.plot(nums, nums/10.)
plt.plot(nums, np.sin(nums))
plt.plot(nums, np.cos(nums))
```

```python
plt.plot(nums, np.power(10, 10*nums))
plt.grid(True)
```

```python
plt.semilogy(nums, np.power(10, 10*nums))
plt.grid(True)
```

```python
# for both axes
plt.loglog(nums, np.power(10, 10*nums))
plt.grid(True)
```

```python
plt.plot(nums, np.power(10, 10*nums)-100)
plt.gca().set_yscale('log')
plt.grid(True)
```

```python
plt.plot(nums, np.power(10, 10*nums)-100)
plt.gca().set_yscale('symlog')
plt.grid(True)
```
